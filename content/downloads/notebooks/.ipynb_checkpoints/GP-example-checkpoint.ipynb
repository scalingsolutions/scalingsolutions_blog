In our previous posting we talked about SVR and in particular GP. These two machines have a lot of features in common, but are usually considered different. I hope ot make the connection between them more concrete using a couple examples. In this post we'll explore the Gaussian processes method. Then we'll explore the SVR. On the way you'll learn both and see how they both live in the same methodological family.

## Gaussian process linear algebra review ##
As a refresher, let's go over the linear algebra of the GP one more time.

We start with a training set $\vec{X},\vec{Y}$ and use the kernel to form the linear equations,
$$ \left( \bar{M} +\epsilon \mathbb{1} \right) \vec{\alpha} = \vec{Y} $$
where $\bar{M}$ is the matrix formed by the kernel, $\epsilon\geq0$ is a regularizer, $\mathbb{1}$ is the identity matrix, and $\vec{\alpha}$ is the vector we are solving for that will be used to form our estimator.

Once we have solved for $\vec{\alpha}$ we form the estimator using,
$$ Y^\star = \vec{\alpha}*\vec{M}^* + \mu $$
where $Y^\star$ is the estimate for a sample point $X^\star$, $\mu$ is the average value of the estimator computed when solving for $\vec{\alpha}$, and $\vec{M}^*$ whose elements are computed by applying the kernel to the training set coordinates and the sample point.

## Gaussian process procedure review ##
To form our GP machine we will need to,
<ol>
<li> Define a training set $\vec{X},\vec{Y}$
<li> Define the kernel $K$ and hyperparameters $\vec{\theta}$
<li> Use $K$ to form the matrix $\bar{M}$
<li> Invert $\bar{M}$ and apply to $\vec{Y}$ to get $\vec{\alpha}$ and $\mu$
<li> Use $\vec{\alpha}$, $K$, $\vec{\theta}$, and $\vec{X}$ to form our estimator
</ol>

We can make a class that will do these things,

```python
import numpy as np
import pprint

class GP:
    
    def __init__(self,X_train,Y_train,kernel,theta,epsilon):
        self.X = X_train
        self.Y = Y_train
        self.Kernel = kernel
        self.Theta = theta
        self.Epsilon = epsilon
        
    def _buildMatrix(self):
        n=len(self.X)
        self.M=np.zeros((n,n))
        for i in range(n):
            self.M[i][i]=1.0+self.Epsilon
            for j in range(i+1,n):
                self.M[i][j]=self.M[j][i]= self.Kernel(self.Theta, self.X[i],self.X[j])
    
    def _computeAlpha(self):
        self._alpha = np.dot( np.linalg.inv(self.M) ,self.Y)
        
    def train(self):
        self._buildMatrix()
        self._computeAlpha()
        
    def estimate(self,X_star):
        m=lambda x: self.Kernel(self.Theta, self.X, x)
        return np.array([np.dot(self._alpha, m(x_star)) for x_star in X_star])
        
    def rmse_training_error(self):
        return  np.sqrt(np.mean((self.estimate(X)-Y)**2))
```

# Clean data #
Let's start by approximating a smooth function, $\cos(x)$, with no noise. This will allow us to see how well the SVR can work in an ideal world.

```python
# We'll start with a pretty small training set
n=5
two_pi = 2*np.pi

# training set
X = np.arange(0,two_pi,two_pi/n)
Y = np.cos(X)

#hyperparameter
theta=0.5
#regularizer
epsilon=0.0

# define the kernel, takes vectors or scalars for all args
def gaussian_kernel(parameter, x1, x2):
    return np.exp(-1.0*parameter*(x1-x2)**2)

# This is how easy it is to use our class
Gaussian_GP = GP(X,Y,gaussian_kernel,theta,epsilon)
Gaussian_GP.train()
print 'RMSE training error:', Gaussian_GP.rmse_training_error()
```

That's it! Now we have a trained regression machine ready to make us predictions. Let's see how well it does. First, lets see how our predictions for the training set compare to the real values. They should be pretty darn close considering that our RMSE, as shown in the last step, is tiny.

```python
import matplotlib.pyplot as plt
Y_est = Gaussian_GP.estimate(X)
plt.scatter(X,Y)
plt.scatter(X,Y_est)

# smooth cos for reference
fine_X = np.arange(0,two_pi,0.01)
fine_Y = np.cos(fine_X)
plt.plot( fine_X, fine_Y)
plt.ylabel("F(X)")
plt.xlabel("X")

plt.show()
```

Yeah, that looks dead on. Now let's try to interpolate using our SVR machine inside and outside of our trinaing data range.

```python
X_test = np.arange(-2.0*np.pi,np.pi,0.1)
Y_test = np.cos(X_test)
Y_estimate = Gaussian_GP.estimate(X_test)

plt.plot( fine_X, fine_Y)
plt.plot(X_test,Y_test)
plt.scatter(X_test,Y_estimate)
plt.ylabel("F(X)")
plt.xlabel("X")
plt.show()
```

Oops! What happened? Well, we left the region where we had training data. If we wanted to fit all the way down to $-2\pi$ then we should have added training data there. You can see that in the region where have training data, the predictor does quite well. The predictor follows through the data pretty well.

```python
plt.plot(X_test,Y_test-Y_estimate)
plt.ylabel("Error")
plt.xlabel("X")
plt.xlim([-1,np.pi])
plt.ylim([-0.05,0.05])
plt.show()
```

## Choosing the kernel ##
We can try other kernels and see how they do. We'll start with a quadratic kernel.

```python
#hyperparameter
theta=1.0
#regularizer
epsilon=0.01

# define the kernel, takes vectors or scalars for all args
def quadratic_kernel(parameter, x1, x2):
    return parameter*(x1-x2)**2

# This is how easy it is to use our class
Quadratic_GP = GP(X,Y,quadratic_kernel,theta,epsilon)
Quadratic_GP.train()
print 'RMSE training error:', Quadratic_GP.rmse_training_error()
```

The RMSE looks a bit high for the training set. Maybe that's ok. We'll plot the estimator and and look at it's error.

```python
X_test = np.arange(-2.0*np.pi,2.0*np.pi,0.1)
Y_test = np.cos(X_test)
Y_estimate = Quadratic_GP.estimate(X_test)

plt.plot( fine_X, fine_Y)
plt.plot(X_test,Y_test)
plt.scatter(X_test,Y_estimate)
plt.ylabel("F(X)")
plt.xlabel("X")
plt.show()
```

Those don't look good. You can see the quadratic kernel forces the estimator to be quadratic as you move out of range of the training set. Also, because the kernel increases in magnitude as the difference between it's points squared, when a test point is estmated, it's value is most strongly determined by the points furthest away! That's no good. We want kernels that are gig when points are nearby.

Let's try another kernel that decays slower than the Gaussian, the Laplacian.

```python
#hyperparameter
theta=1.0
#regularizer
epsilon=0.0

# define the kernel, takes vectors or scalars for all args
def laplacian_kernel(parameter, x1, x2):
    return np.exp(-parameter*np.abs(x1-x2))

# This is how easy it is to use our class
Laplacian_GP = GP(X,Y,laplacian_kernel,theta,epsilon)
Laplacian_GP.train()
print 'RMSE training error:', Laplacian_GP.rmse_training_error()
```

That error looks pretty small, Maybe we have another good function. Let's plot and see...

```python
X_test = np.arange(-2.0*np.pi,2.0*np.pi,0.1)
Y_test = np.cos(X_test)
Y_estimate = Laplacian_GP.estimate(X_test)

plt.plot( fine_X, fine_Y)
plt.plot(X_test,Y_test)
plt.scatter(X_test,Y_estimate)
plt.ylabel("F(X)")
plt.xlabel("X")
plt.show()
```

Oops again! The training set is well taken care of, but the interpolation stinks. We have a cuspy interpolator, that's no good. Notice again that as we move away from the training data, the estimator decays like a Laplacian.

So, we need a smooth function with no cusps. The Gaussian works well, but, maybe for this function there's something better we can use. Let's try something periodic like cosine.

```python
#hyperparameter
theta=1.0
#regularizer
epsilon=0.0

# define the kernel, takes vectors or scalars for all args
def cosine_kernel(parameter, x1, x2):
    return np.cos(parameter*np.abs(x1-x2))

# This is how easy it is to use our class
cos_GP = GP(X,Y,cosine_kernel,theta,epsilon)
cos_GP.train()
print 'RMSE training error:', cos_GP.rmse_training_error()
```

```python
X_test = np.arange(-2.0*np.pi,2.0*np.pi,0.1)
Y_test = np.cos(X_test)
Y_estimate = (1.0/np.pi)*cos_GP.estimate(X_test)

plt.plot( fine_X, fine_Y)
plt.plot(X_test,Y_test)
plt.scatter(X_test,Y_estimate)
plt.ylabel("F(X)")
plt.xlabel("X")
plt.show()
```

Once again, no surprises here. We use a kernel that looks the same as thefunction we're trying to interpolate and we get something reasonable out. There's a slight shift in the output due to my omission of a phase factor in the kernel. I also stuck in a normalization factor of $1/\pi$ by hand. Both could have been optimized.

The take-hme point is that, unless you have a good idea for what the function should be, you should probably use a Gaussian kernel. If you have special knowledge of some asymptotics then you can be clever. Otherwise, it's probably best to use what works.

## Choosing $\epsilon$ ##
We can compute the RMSE for a fine grid of test points and see how well, we're really doing. Let's go ahead and see how it varies as we change the size of the training set too.

```python
sample_sizes=range(1,42,2)
training_error_list=[]
testing_error_list=[]
epsilon=0
for n in sample_sizes:
    X = np.arange(0,two_pi,two_pi/n)
    Y = np.cos(X)
    Gaussian_GP = GP(X,Y,gaussian_kernel,theta,epsilon)
    Gaussian_GP.train()
    training_error_list.append( Gaussian_GP.rmse_training_error() )
    test_RMSE = np.sqrt(np.mean((Gaussian_GP.estimate(fine_X)-fine_Y)**2))
    testing_error_list.append(test_RMSE)
    
Ptrain = plt.semilogy(sample_sizes,training_error_list,ls='--')
Ptest  = plt.semilogy(sample_sizes,testing_error_list)
plt.legend([Ptrain[0],Ptest[0]],["Training","Testing"],loc='center left')
plt.ylabel("RMSE Error")
plt.xlabel("Training set size")
plt.show()
```

That seems odd, right? As we increase the size of the trinaing set, the error on our predictions for points in the trinaing set increases. Why is that? Well, it's because the matrix we're inverting is becoming more and more singular. Finally, when we get to 20 points in the training set, we can't do any better. How could we try to fix that? Let's add the regularizer!

```python
sample_sizes=range(1,42,2)
epsilon_range = np.arange(-16,0,2)
n_eps=len(epsilon_range)
Ptrain=[]
Ptest=[]
Ttest=[]

for eps_i,eps_exp in enumerate(epsilon_range):
    epsilon=10**eps_exp
    training_error_list=[]
    testing_error_list=[]
    for n in sample_sizes:
        X = np.arange(0,two_pi,two_pi/n)
        Y = np.cos(X)
        Gaussian_GP = GP(X,Y,gaussian_kernel,theta,epsilon)
        Gaussian_GP.train()
        training_error_list.append( Gaussian_GP.rmse_training_error() )
        test_RMSE = np.sqrt(np.mean((Gaussian_GP.estimate(fine_X)-fine_Y)**2))
        testing_error_list.append(test_RMSE)
    
#     like a heat map, when eps is small the line plot is blue, large is red 
    c_frac=1.0*eps_i/n_eps
    Ptrain.append( plt.semilogy(sample_sizes,training_error_list, ls="--", c=[c_frac,0,1.0-c_frac])[0] )
    Ttest.append("$\epsilon=10^{%s}$"%(eps_exp))
    Ptest.append( plt.semilogy(sample_sizes,testing_error_list, c=[c_frac,0,1.0-c_frac])[0] )
plt.xlim([0,42])
L1=plt.legend([Ptrain[0],Ptest[0]],["Training","Testing"],loc='lower right')
plt.legend(Ptest,Ttest,bbox_to_anchor=[1.01,-0.1,0.3,1.0])
plt.ylabel("RMSE Error")
plt.xlabel("Training set size")
plt.gca().add_artist(L1)
plt.show()
```

We can see from the preceding plot that as we increase the regularizer we increase the error in our predictions for the training set and the test set. But, we also get to increase the size of the training set. For a noisy function we can have problems with overfitting that manifests itself the same way.

In our case, this is an artifact of the way we are choosing our hyperparameter. If we were optimizing it, this wouldn't be an issue. Let's go ahead and show what happens ot the error as we scan through $\theta$ values with a fixed $\epsilon=0$.

```python
sample_sizes=range(1,42,2)
theta_range = np.arange(0.25,6,0.25)
n_theta=len(theta_range)
Ptrain=[]
Ptest=[]
Ttest=[]
epsilon=0
for theta_i,theta in enumerate(theta_range):
    training_error_list=[]
    testing_error_list=[]
    for n in sample_sizes:
        X = np.arange(0,two_pi,two_pi/n)
        Y = np.cos(X)
        Gaussian_GP = GP(X,Y,gaussian_kernel,theta,epsilon)
        Gaussian_GP.train()
        training_error_list.append( Gaussian_GP.rmse_training_error() )
        test_RMSE = np.sqrt(np.mean((Gaussian_GP.estimate(fine_X)-fine_Y)**2))
        testing_error_list.append(test_RMSE)
    
#     like a heat map, when theta is small the line plot is blue, large is red 
    c_frac=1.0*theta_i/n_theta
    Ptrain.append( plt.semilogy(sample_sizes,training_error_list, ls="--", c=[c_frac,0,1.0-c_frac])[0] )
    Ttest.append("$\\theta=%s$"%(theta))
    Ptest.append( plt.semilogy(sample_sizes,testing_error_list, c=[c_frac,0,1.0-c_frac])[0] )
plt.xlim([0,42])
L1=plt.legend([Ptrain[0],Ptest[0]],["Training","Testing"],loc='lower right')
plt.ylabel("RMSE Error")
plt.xlabel("Training set size")
plt.legend(Ptest,Ttest,bbox_to_anchor=[1.41,0.0,0.3,1.0],ncol=2)
plt.gca().add_artist(L1)
plt.show()
```

Here we can see that there is an optimal $\theta$ for a given size of the test set. For a given training set size you can find the optimal $\theta$ by finding the test set with the lowest error. In this case the lowest error is of the order $10^{-4}$ across these sample sizes.

## Noisy Data ##
Great! Now we have a pretty good idea of what all the inputs to the GP do. Let's apply it to some noisy data and see what happens. What we'll do is take our cosine function and add Gaussian random noise into it. Using this noisy training set we will train a GP and use it to estimate the function. What we want to know is, by increasing the regularizer, can we get estimates of the real function which are more accurate than the noisy estimator we used to build the GP.

```python
# reset some parameters to something reasonable
n=20
X = np.arange(0,two_pi,two_pi/n)
theta=0.5

# for book keeping
Ptrain=[]
Ptest=[]
Popt=[]
Ttest=[]

# here is the range of values for the log(noise)
noise_range=np.arange(-2,0.5,0.5)

# here is the range of values for the log(epsilon)
epsilon_range = np.arange(-4,0.5,1)
n_eps=len(epsilon_range)

for eps_i,eps_exp in enumerate(epsilon_range):
    epsilon=10**eps_exp
    train_error=[]
    test_error=[]
    for noise_factor in noise_range:
        noise = 10**noise_factor
        testing_error_list=[]
        training_error_list=[]
        n_samples=40
        for _ in range(n_samples): #average over a bunch of different random noise configs
            Y = np.cos(X)+noise*np.random.normal(size=n)
            Gaussian_GP = GP(X,Y,gaussian_kernel,theta,epsilon)
            Gaussian_GP.train()
            training_error_list.append( Gaussian_GP.rmse_training_error() )
            test_RMSE = np.sqrt(np.mean((Gaussian_GP.estimate(fine_X)-fine_Y)**2))
            testing_error_list.append(test_RMSE)
        test_error.append(np.mean(testing_error_list))
        train_error.append(np.mean(training_error_list))
    c_frac=1.0*eps_i/n_eps
    Ptest.append( plt.plot(noise_range,np.log(test_error),c=[c_frac,0,1.0-c_frac])[0] )
    Ptrain.append( plt.plot(noise_range,np.log(train_error),c=[c_frac,0,1.0-c_frac],ls='--')[0] )
    Ttest.append("$10^{%s}$"%(noise_factor))
plt.xlabel("$\log$ noise")
plt.ylabel("$\log$ RMSE error")
plt.show()
```

As we increase the noise in our training set, $\vec{Y}$, we increase the error in our GP estimator. We can see that as the noise increases, so too does the optimal regularization parameter, $\epsilon$. In fact, the optimal regularization is related to the variance, $\sigma^2$, of the noise and the hyperparameter.

```python
n=20
X = np.arange(0,two_pi,two_pi/n)
theta=0.1

Ptrain=[]
Ptest=[]
Popt=[]
Ttest=[]

noise_range=np.arange(-3,0.5,0.25)
n_noise=len(noise_range)

train_error=[]
test_error=[]
for noise_i, noise_factor in enumerate(noise_range):
    noise = 10**noise_factor
    epsilon = noise**2 # here we set epsilon to the variance of the noise
    testing_error_list=[]
    training_error_list=[]
    n_samples=80
    for _ in range(n_samples):
        Y = np.cos(X)+noise*np.random.normal(size=n)
        Gaussian_GP = GP(X,Y,gaussian_kernel,theta,epsilon)
        Gaussian_GP.train()
        training_error_list.append( Gaussian_GP.rmse_training_error() )
        test_RMSE = np.sqrt(np.mean((Gaussian_GP.estimate(fine_X)-fine_Y)**2))
        testing_error_list.append(test_RMSE)
    test_error.append(np.mean(testing_error_list))
    train_error.append(np.mean(training_error_list))
plt.plot(noise_range,np.log(test_error))
plt.plot(noise_range,np.log(train_error),ls="--")

plt.xlabel("$\log$ noise")
plt.ylabel("$\log$ RMSE error")
plt.show()
```

If we optimize the parameter $\theta$ we can do better.
